```python
import sqlite3
```

```python
import pandas as pd
```

```python
con = sqlite3.connect('sakila.db')
```

```python
def sql_to_df(sql_query):
    df = pd.read_sql(sql_query, con)
    return df
```

```python
query = '''select first_name, last_name
                    from  customer'''
sql_to_df(query)
```

```python
query = 'select * from customer'
sql_to_df(query)
```

```python
sql_to_df('select country_id from city')
```

```python
sql_to_df('select distinct(country_id) from city')
```

```python
query = '''select * 
                    from customer
                    where store_id = 1'''
sql_to_df(query).head()
```

```python
query = '''select * 
                    from customer
                    where first_name = 'MARY'
                    '''
sql_to_df(query).head()
```

```python
query = '''select * from film
                    where release_year = 2006 and rating='R'
                    '''
sql_to_df(query)
```

```python
query = '''select * from film
                    where rating='PG' or rating='R'
                    '''
sql_to_df(query)
```
